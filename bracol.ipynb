```python
import shutil
import glob
files = glob.glob('val/*.png')
origin = r'C:\\Users\\User\\Documents\\TCC\\bracol-edit\\coffee-datasets\\segmentation\\images\\val'
final = r'C:\\Users\\User\\Documents\\TCC\\bracol-edit\\coffee-datasets\\segmentation\\annotations\\copiasVal'

file_id = 0
anot_id = 0
for file in files:
  print(file)
  new = file[4:].replace('_mask', '').replace('.png', '.jpg')
  print(new)
  id = int(new.replace('.jpg', ''))
  if id in novo.index:
    original = cv.imread(file)
    red = np.array([0,0,255])
    mask = cv.inRange(original, red, red)

    contours = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)[0]
    shutil.copy(origin + f'\\{new}', final + f"\\{new}")
    data["images"].append({
      "id": file_id,
      "file_name": f'{new}',
      "license": 1,
      "date_captured": "2022-07-21T03:05:36+00:00",
      "width": 2048,
      "height": 1024
    })
    if len(contours) > 0:
      for contour in contours:
        data['annotations'].append(
          {
            "id": anot_id,
            "image_id": file_id,
            "category_id": int(np.where(novo.loc[id].values == 1)[0][0] + 1),
            "segmentation": [contour.flatten().tolist()],
            "bbox": cv.boundingRect(contour),
            "area": cv.contourArea(contour),
            "iscrowd": 0
          }
        )
        anot_id += 1
    file_id+=1
```
